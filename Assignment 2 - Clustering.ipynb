# Performance Analysis

1. Create a dataset : 700 instances data points and each instances belongs to 4 clusters

```python
from sklearn.datasets import make_blobs
X, cluster_assignments = make_blobs(n_samples=700, centers=4,
                                    cluster_std=0.60, random_state=0)
X
```

* cluster_assignments : it is the true clusters(array)
* X : the data points of our dataset

```python
cluster_assignments
```

## Compare MyKMeans result with real cluster

* import mine version of KMeans and compare the performances with real clusters

```python
from kmeans import KMeans

my_kmeans = KMeans(k=4, max_iterations = 1000)

my_clusters, my_centroids = my_kmeans.fit(X)
```

```python
my_clusters
```

* my_clusters : the cluster result from my kmeans
* my_centroids : the centroids of each cluster (in this case number of centroids will be 4 )

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

centroids = np.array([X[cluster_assignments == i].mean(axis=0) for i in range(4)])

# True clusters
ax[0].scatter(X[:, 0], X[:, 1], c=cluster_assignments, s=50, cmap='viridis', label='True Clusters')
ax[0].set_title('True Cluster Assignments')
ax[0].legend()

# My KMeans clusters
ax[1].scatter(X[:, 0], X[:, 1], c=my_clusters, s=50, cmap='viridis', label='My KMeans Clusters')
ax[1].set_title('My KMeans Cluster Assignments')
ax[1].legend()


for i, centroid in enumerate(centroids):
    ax[0].text(centroid[0], centroid[1], str(i), color='orange', fontsize=20)
for i, centroid in enumerate(my_centroids):
    ax[1].text(centroid[0], centroid[1], str(i), color='orange', fontsize=20)


plt.tight_layout()
plt.show()
```

```python
cluster_dic = {
    3:0,
    2:2,
    0:1,
    1:3}


match_count = 0
for i in range(700):
    if my_clusters[i] == cluster_dic.get(cluster_assignments[i]):
        match_count +=1

accuracy = match_count / 700
print("Matching Number: ",match_count)
print("Performance of My KMeans Algorithm, Measured by Match Rate:",accuracy)
```

### Rand Index & Adjusted Rand Index
* Rand Index -  Another commonly used metric is the Rand Index. It computes a similarity measure between two clusters by considering all pairs of samples and counting pairs that are assigned in the same or different clusters in the predicted and true clusterings.
  * The formula of the Rand Index is:
      * RI = number of agreeing pairs / number of pairs
* Adjusted Rand Index - Rand index adjusted for chance.
    * The Rand Index computes a similarity measure between two clusterings by considering all pairs of samples and counting pairs that are assigned in the same or different clusters in the predicted and true clusterings.
    * The raw RI score is then adjusted for chance into the ARI score using the following scheme:
        * ARI = (RI - Expected RI) / (MAX(RI) -Expected RI)

* Rand Index : between 0 and 1, 1.0 stands for perfect match.
* Adjusted Rand Index: Similarity score between -0.5 and 1.0. Random labelings have an ARI close to 0.0. 1.0 stands for perfect match.

```python
from sklearn.metrics import rand_score, adjusted_rand_score
my_rand_score = rand_score(cluster_assignments, my_clusters)
my_ari = adjusted_rand_score(cluster_assignments, my_clusters)

print("Rand Index : ",my_rand_score)
print("Adjusted Rand Index : ",my_ari)
```

## Compare scikit-learn result with real cluster

```python
from sklearn.cluster import KMeans

sk_kmeans = KMeans(n_clusters=4)
sk_kmeans.fit(X)

sk_kmeans.labels_
```

```python
sk_kmeans.cluster_centers_
```

* sk_kmeans.labels_ : the cluster result from sklearn* sk_kmeans.cluster_centers_s : the centroids of each cluster (in this case number of centroids will be 4) from sklearn)

```python

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

centroids = np.array([X[cluster_assignments == i].mean(axis=0) for i in range(4)])

# True clusters
ax[0].scatter(X[:, 0], X[:, 1], c=cluster_assignments, s=50, cmap='viridis', label='True Clusters')
ax[0].set_title('True Cluster Assignments')
ax[0].legend()

# My KMeans clusters
ax[1].scatter(X[:, 0], X[:, 1], c=sk_kmeans.labels_, s=50, cmap='viridis', label='Scikit-learn Clusters')
ax[1].set_title('scikit-learn Assignments')
ax[1].legend()


for i, centroid in enumerate(centroids):
    ax[0].text(centroid[0], centroid[1], str(i), color='orange', fontsize=20)
for i, centroid in enumerate(sk_kmeans.cluster_centers_):
    ax[1].text(centroid[0], centroid[1], str(i), color='orange', fontsize=20)


plt.tight_layout()
plt.show()
```

```python
cluster_dic = {
    3:1,
    2:0,
    0:3,
    1:2}


match_count = 0
for i in range(700):
    if sk_kmeans.labels_[i] == cluster_dic.get(cluster_assignments[i]):
        match_count +=1

accuracy = match_count / 700
print("Matching Number: ",match_count)
print("Performance of Scikit-learn Algorithm, Measured by Match Rate:",accuracy)
```
